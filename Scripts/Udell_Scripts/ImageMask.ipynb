```python
import pandas as pd
import cv2

# Get the id numbers from the CSV
dataset = pd.read_csv(r"C:\Users\moin2\OneDrive - University of Georgia\Spring 2023\Silene\Light Petals\LightPetal_id.csv")
ids = dataset["id"]

# This is unnecessary, but you probably want to see how the the program is progressing, so we'll capture the number of 
# photos that need to be processed and start a counter at 1
total = len(ids)
counter = 1

for i in ids:
    
    # Define the path to the image and load it
    image = cv2.imread(r"C:\Users\moin2\OneDrive - University of Georgia\Spring 2023\Silene\Light Petals" + "\\" + "AllPhotos" + str(i) + ".jpg")
    
    # Define the path to the mask and load it
    mask = cv2.imread(r"C:\Users\moin2\OneDrive - University of Georgia\Spring 2023\Silene\Light Petals" + "\\" + "AllPhotos" + str(i) + "_Simple Segmentation.jpeg")
    
    print(image.shape)
    print(mask.shape)
    
    # Apply the mask to the image
    segmented = cv2.bitwise_or(mask, image)
    
    # Save the resulting image
    cv2.imwrite(r"C:\Users\moin2\OneDrive - University of Georgia\Spring 2023\Silene\Light Petals" + "\\" + "segmented" + str(i) + ".jpeg", segmented)
    
    # Print a message with the number done out of the total number that need to be completed
    print(str(counter) + "/" + str(total))
    # increase the counter by 1
    counter = counter + 1
```
